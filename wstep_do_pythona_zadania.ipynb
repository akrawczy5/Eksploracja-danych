ZADANIE 1: Analiza danych sprzedażowych
---------------------------------------
Dane: Miesięczna sprzedaż produktu w PLN

sprzedaz = [12500, 13200, 15800, 14200, 16500, 18200, 17800, 19200,
            20500, 21300, 23500, 25200]

a) Oblicz podstawowe statystyki (średnia, mediana, min, max)

b) Znajdź miesiące ze sprzedażą powyżej średniej

c) Oblicz procentowy wzrost sprzedaży (grudzień vs styczeń)

d) Stwórz wykres liniowy pokazujący trend

e) Dodaj linię średniej na wykresie


````python
sprzedaz = [12500, 13200, 15800, 14200, 16500, 18200, 17800, 19200, 20500, 21300, 23500, 25200]
miesiace = ["styczeń", "luty", "marzec", "kwiecień", "maj", "czerwiec", "lipiec", "sierpień", "wrzesień", "październik", "listopad", "grudzień"]
print("\nMiesiące i sprzedaz:")
for miesiac, sprzedaz_value in zip(miesiace[:12], sprzedaz[:12]):
    print(f"  {miesiac}: {sprzedaz_value} zł")

print(f"\nMin sprzedaz: {min(sprzedaz)} zł")
print(f"Max sprzedaz: {max(sprzedaz)} zł")
print(f"Średnia: {round(np.mean(sprzedaz), 2)} zł")
print(f"Mediana: {np.median(sprzedaz)} zł")

print("\nMiesiące ponad średnią:")
for miesiac, sprzedaz_value in zip(miesiace, sprzedaz):
    if sprzedaz_value > np.mean(sprzedaz):
        print(f"  {miesiac}: {sprzedaz_value} zł")

wzrost_procentowy = ((sprzedaz[-1] - sprzedaz[0]) / sprzedaz[0]) * 100
print(f"\nProcentowy wzrost sprzedaży (grudzień vs styczeń): {round(wzrost_procentowy, 2)}%")

plt.subplots()
plt.plot(miesiace, sprzedaz, label='Sprzedaż')
plt.axhline(np.mean(sprzedaz), color='r', linestyle='--', label=f'Średnia: {round(np.mean(sprzedaz), 2)} zł')
plt.xlabel('Miesiące')
plt.ylabel('Sprzedaż w PLN')
plt.title('Zależność sprzedaży od miesiąca')
plt.grid(True, alpha=0.5)
plt.legend()
plt.tight_layout()
plt.show()
````

ADANIE 2: Słowniki - Analiza portfolio
---------------------------------------
Stwórz słownik opisujący portfolio inwestycyjne:
```
portfolio = {
    'AAPL': {'liczba': 10, 'cena_zakupu': 150, 'cena_aktualna': 175},
    'GOOGL': {'liczba': 5, 'cena_zakupu': 2800, 'cena_aktualna': 2950},
    'MSFT': {'liczba': 15, 'cena_zakupu': 300, 'cena_aktualna': 320}
}
```

a) Oblicz całkowitą wartość początkową portfolio

b) Oblicz aktualną wartość portfolio

c) Oblicz zysk/stratę dla każdej akcji

d) Które akcje przynoszą największy zwrot (w %)?


````python
portfolio = {
    'AAPL': {'liczba': 10, 'cena_zakupu': 150, 'cena_aktualna': 175},
    'GOOGL': {'liczba': 5, 'cena_zakupu': 2800, 'cena_aktualna': 2950},
    'MSFT': {'liczba': 15, 'cena_zakupu': 300, 'cena_aktualna': 320}
}

wartosc_poczatkowa = sum(dane['liczba'] * dane['cena_zakupu'] for dane in portfolio.values())
print("Całkowita wartość początkowa portfolio:", wartosc_poczatkowa)

wartosc_aktualna = sum(dane['liczba'] * dane['cena_aktualna'] for dane in portfolio.values())
print("Całkowita wartość aktualna portfolio:", wartosc_aktualna)

zyski_straty = {}

for item, dane in portfolio.items():
    zysk_strata = dane['liczba'] * dane['cena_aktualna'] - dane['liczba'] * dane['cena_zakupu']
    zyski_straty[item] = zysk_strata

print("\nZysk/strata dla:")
for item, wartosc in zyski_straty.items():
    print(f"{item}: {wartosc}")

zwroty_proc = {}

for item, dane in portfolio.items():
    zwrot = ((dane['cena_aktualna'] - dane['cena_zakupu']) / dane['cena_zakupu']) * 100
    zwroty_proc[item] = zwrot

print("\nZwrot procentowy dla każdej akcji:")
for item, zwrot in zwroty_proc.items():
    print(f"{item}: {zwrot:.2f}%")

najlepsza = max(zwroty_proc, key=zwroty_proc.get)
print(f"\nNajwiększy zwrot przynosi: {najlepsza} ({zwroty_proc[najlepsza]:.2f}%)")
````

ZADANIE 4: Symulacja Monte Carlo
---------------------------------
Zasymuluj 1000 rzutów kostką (6-ścienną):

a) Wygeneruj dane używając np.random.randint(1, 7, 1000)

b) Stwórz histogram pokazujący częstość każdej liczby

c) Oblicz teoretyczne i empiryczne prawdopodobieństwo każdej liczby

d) Czy wyniki są zgodne z teorią? (każda liczba ~16.67%)

````python
rzuty = []

rzuty.append(np.random.randint(1, 7, 1000))

plt.hist(rzuty, bins=30, edgecolor='black')
plt.title("Symulacja Monte Carlo - rzuty kostką")
plt.xlabel("Wartości na kostce")
plt.ylabel("Częstość")
plt.show()

unikalne, liczby = np.unique(rzuty, return_counts=True)
p_emp = liczby / 1000
p_teor = np.full(6, 1/6)

for i in range(6):
    print(f"Liczba {i+1}: prawdopodobieństwp empiryczne={p_emp[i]:.5f}, prawdopodobieństwo teoretyczne={p_teor[i]:.5f}, różnica={p_emp[i]-p_teor[i]:.5f}")
    if p_emp[i]-p_teor[i] != 0:
      print("Wynik niezgodny z teorią")
    else:
      print("Wynik zgodny z teorią")
````
